## Równanie Lotki-Volterry
Nieliniowy układ równań różniczkowych pierwszego stopnia opisujący relację drapieżnik-ofiara:

$$\frac{\delta x}{\delta t} = (a - by)x,$$

$$\frac{\delta y}{\delta t} = (cx - d)y.$$

Na rozwiązanie składają się trzy programy: generator zbioru danych treningowych i testowych, trenowanie sieci neuronowej oraz testowanie z wyświetlaniem efektów estymowania.

```python
# IMPORT WYKORZYSTYWANYCH BIBLIOTEK

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import tensorflow as tf
import random
```

W pliku *data_generator.py* na przedziale [0,50] podzielonym na 1000 części generuję zbiór danych treningowych i testowych: przy mojej sieci pierwszy składa się z 20000 rekordów, drugi natomiast z 200.
Po zdefiniowaniu równania różczniczkowego Lotki-Volterry jako funkcji rozwiązuję je przy pomocy *solve_ivp* z wylosowanymi punktami startowymi na przedziale (1,20). Następnie dopisuję do odpowiedniego pliku ostatnie 200 punktów rozwiązania funkcji $x$ i $y$.

Zadaniem sieci będzie predykcja stałych $a,b,c,d$ z równania różniczkowego, więc każdy generowany przykład będzie składał się z jednej czterowymiarowej zmiennej objaśnianej i 400 zmiennych objaśniających.

```python
# ZDEFINIOWANIE PRZEDZIAŁU ROZWĄZANIA

start = 0
end = 50
partition = 1000
```

```python
# ROZWIAZYWANIE ROWNANIA ROZNICZKOWEGO

def lotkavolterra(t, z, a, b, c, d):
    x, y = z
    return [a*x - b*x*y, c*x*y - d*y]

def preparing_lotka_data(a,b,c,d):
    solution = solve_ivp(lotkavolterra,
                    [start, end], # przedział czasowy
                    [random.randint(1,20),random.randint(1,20)], # punkty startowe
                    args=(a, b, c, d), #parametry równania Lotki-Voltery
                    dense_output=True)

    t = np.linspace(start, end, partition)
    z = solution.sol(t)

    x = [z.T[800:1000]]

    res_x = [0]*400

    for i in range(200):
        res_x[i] = x[0][i][0]
        res_x[200+i] = x[0][i][1]

    res_y = [a,b,c,d]

    return res_x, res_y
```

```python
# STWORZENIE SIECI NEURONOWEJ

model = tf.keras.Sequential([
    tf.keras.layers.Dense(600, input_dim = 400, activation = "sigmoid"),
    tf.keras.layers.Dense(400, activation="tanh"),
    tf.keras.layers.Dense(300, activation = "softplus"),
    tf.keras.layers.Dense(200, activation = "sigmoid"),
    tf.keras.layers.Dense(100, activation = "tanh"),
    tf.keras.layers.Dense(4, activation="softplus")
])

model.compile(optimizer = "adamax", loss ="MSE")

# model.fit(x_train, y_train, epochs = 200)

# model.save_weights("model1_weights.h5")
```

```python
# ZDEFINIOWANIE FUNKCJI DAJACEJ PRZYKLAD PREDYKCJI

def random_example():

    index = random.randint(0,199)

    x = x_test[index]
    x = x.reshape((2,200))
    t1 = np.linspace(40, 50, 200)
    t2 = np.linspace(50, 60, 200)

    solution = solve_ivp(lotkavolterra,
                        [50, 60], # przedział czasowy
                        [x[0][199],x[1][199]], # punkty startowe
                        args=(y_predict[index]), #parametry równania Lotki-Voltery
                        dense_output=True)

    z = solution.sol(t2)


    plt.plot(t1, x[0])
    plt.plot(t1, x[1])
    plt.plot(t2,z.T, linestyle = "dashed")
    plt.xlabel('Czas')
    plt.legend(['Prey', 'Predator','predicted Prey', 'predicted Predator'], shadow=True)

    plt.title('Równanie Lotki-Voltery')
    plt.show()
```

```python
# LOSOWO WYGENEROWANE PRZYKŁADY

examples_num = 5

for i in range(examples_num):
    random_example()
```
